```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
os.chdir('/home/sidrah19220/nmd/loomfiles/updates/point4')
```

```python
mtx = pd.read_csv('Out_data_eud.csv', index_col =0)
mtx = mtx[mtx['Expression Diffrence from rank1']==0]
#unique_mat = mtx.drop_duplicates(subset = 'Mature Cell ID', keep ='first')
unique_mat = mtx.drop_duplicates(subset = 'Immature/Transition Cell ID', keep ='first')
l = unique_mat['Immature/Transition Cell ID'].values.tolist()
```

```python
tran =0
imm = 0
for k in l:
    if k.startswith('IMT'):
        tran = tran+1
    if k.startswith('IMMA'):
        imm = imm +1
```

```python
li = [tran, imm]
```

```python
li
```

```python
x = ['IMT CELLS', 'IMMATURE CELLS']
plt.bar(x, li)
plt.title('Count of each cell type')
plt.xlabel('Nearest Cells')
plt.ylabel('# of paired')
plt.savefig('Point4-count-euc.pdf')
```

```python
li = [(tran/214)*100, (imm/261)*100]
li
```

```python
x = ['IMT CELLS', 'IMMATURE CELLS']
plt.bar(x, li)
plt.title('percentage of count of each cell type')
plt.yticks([0,20, 40, 60, 80, 100])
plt.xlabel('Nearest Cells')
plt.ylabel('# of pairs')
plt.savefig('Point4-percentage-euc.pdf')
```
